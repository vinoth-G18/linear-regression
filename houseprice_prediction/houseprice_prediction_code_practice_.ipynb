BEFORE USING GRADIENT DESCENT TECHNIQUE ,ONLY GIVE RANDOM W,B INPUT FROM USER

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_train=[0,2]
y_train=[300,500]
```

```python
plt.scatter(X_train,y_train,marker='x',c='r')
plt.title("house price")
plt.ylabel("price in dollar")
plt.xlabel("size in sqr feet")
plt.show()
```

MODEL

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

INPUT

```python
w=100
b=300
```

AFTER THE MODEL PLOT

```python

X_train=np.array(X_train)
tmp_f_wb= compute_model_output(X_train,w,b)

plt.plot(X_train,tmp_f_wb,c='b',label='our prediction')
plt.scatter(X_train,y_train,marker='x',c='r',label='actual value')
plt.xlabel('size in sqr feet')
plt.ylabel('price in dollar')

plt.show()
```

PERFECTLY MATCHING AFTER THE MANY ATTEMPT TO GIVE A RANDOM VALUE FOR W AND B

```python
X_test=np.array([1,2,67])
y_test_pred=compute_model_output(X_test,w,b)
print(y_test_pred)
```

AFTER USING GRADIENT DESCENT IN MODEL

cost function


```python
def cost_function(X,y,w,b):
    m=len(X)
    cost=np.sum((w*X+b-y)**2)/(2*m)
    return cost


cost_value=cost_function(X_train,y_train,w,b)
print('j(w,b)=',cost_value)

```

```python
def compute_gradient(X,y,w,b):
    m=len(X)
    dj_dw=np.sum((w*X+b-y)*X) /m
    dj_db=np.sum(w*X+b-y)/m
    return dj_dw,dj_db

def gradient_descent(X,y,w,b,alpha,iterations):
    cost_history=[]
    for i in range(iterations):
        dj_dw,dj_db=compute_gradient(X,y,w,b)
        w-=alpha*dj_dw
        b-=alpha*dj_db


        cost=cost_function(X,y,w,b)
        cost_history.append(cost)


        if i%100==0:
           print('cost',cost,'w',w,'b',b)
    return w,b,cost_history




```

INITIAL VALUES FOR W,B,ALPHA,ITERATION

```python
w = 0
b = 0
alpha = 0.01
iterations = 6000

w_final,b_final,cost_history=gradient_descent(X_train,y_train,w,b,alpha,iterations)
```
