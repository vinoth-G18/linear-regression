```python
import scipy.stats as stats
z_score=stats.zscore(df_numeric)
threshold=3
print("before removing the outliers :",df.shape)
outlier_df=df[(z_score>threshold).any(axis=1)]
df=df[(z_score<=threshold).all(axis=1)]
print("after removing the outliers:",df.shape)

```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(df_numeric)
df_scaled=pd.DataFrame(scalar.transform(df_numeric),columns=df_numeric.columns)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

```

```python
x=df_numeric.drop("price",axis=1)
y=df_numeric["price"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(f"xtrain {x_train.shape}")
print(f"xtest {x_test.shape}")
```

```python
models=[
    
    ("linear regression",LinearRegression())
    
    
   


]
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
for name,model in models:
    print(name)
    print()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print("mean_absolute_error:",mean_absolute_error(y_test,y_pred))
    print("\n")
    print("mean_squared_error:",mean_squared_error(y_test,y_pred))
    print("\n")
    print("r2_score:",r2_score(y_test,y_pred))
    print("\n")
  
    
```

```python
import joblib

joblib.dump(model, "house_price_model.pkl")  # Save model
```
